```python
import pandas as pd
```

We modify the dataset to be able to train on it.

```python
df=pd.read_csv("../data/Wilson_disease_dataset.csv")
df.drop(columns=['Socioeconomic Status'], inplace=True)
df.drop(columns=['Name','Region'],inplace=True)
df.drop(columns=["ATB7B Gene Mutation"], inplace=True)
df.drop(columns=["Kayser-Fleischer Rings"], inplace=True )
df.drop(columns=["Age"], inplace=True)
#leak featureS that almost always predict the presence of the disease
df.to_csv("../data/Wilson_disease_dataset.csv",index=False)
print(df.head())
```

```python
df=pd.read_csv("../data/Wilson_disease_dataset.csv")
columns=['Sex','Alcohol Use']
df=pd.get_dummies(data=df,prefix=columns,columns=columns)
df.to_csv("../data/Wilson_disease_dataset.csv",index=False)
print(df.head())
```

```python
df=pd.read_csv("../data/Wilson_disease_dataset.csv")
coloane=['Sex_Female','Sex_Male','Socioeconomic Status_High','Socioeconomic Status_Low','Socioeconomic Status_Medium','Alcohol Use_False','Alcohol Use_True']
df[coloane]= df[coloane].astype('Int64')
df.to_csv("../data/Wilson_disease_dataset.csv",index=False)
print(df.head())
```

We then put the results as the last column

```python
df=pd.read_csv("../data/Wilson_disease_dataset.csv")
result=df.pop('Is_Wilson_Disease')
df['Is_Wilson_Disease']=result
df.to_csv("../data/Wilson_disease_dataset.csv",index=False)
```

We then also create the simpler dataset that is made out of normal lab tests 

```python
df=pd.read_csv("../data/Wilson_disease_dataset.csv")
df_simplified=df.copy()
df_simplified.drop(columns=['Ceruloplasmin Level','Copper in Blood Serum','Free Copper in Blood Serum','Copper in Urine','Cognitive Function Score','Neurological Symptoms Score'], inplace=True)
df_simplified.to_csv("../data/Wilson_disease_dataset_simple.csv", index=False)
```
